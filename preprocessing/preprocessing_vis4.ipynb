```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../dataset/data_cleaned.csv")
```

```python
df["start_date"] = df["start_date"].str[5:7].astype(int)
```

```python
df
```

```python
ser = df.groupby(["subscription_type", "start_date"])["id"].count()
ser
```

```python
res_dict = {"Customer":[], "Subscriber": []}
for i in res_dict:
    temp = ser[i]
    for j in temp:
        res_dict[i].append(j)
res = pd.DataFrame(res_dict)
```

```python
res.index = range(1, 13)
res
```

```python
res.to_csv("../dataset/data_cleaned_vis4.csv")
```
