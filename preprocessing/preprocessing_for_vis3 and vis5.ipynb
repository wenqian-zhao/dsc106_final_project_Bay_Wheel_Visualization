```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../dataset/data_cleaned.csv")
```

```python
df = df[["start_date_time", "start_station_name"]]
```

```python
df["start_date_time"] = df["start_date_time"].str[:2]
```

```python
sr = df.groupby(["start_date_time", "start_station_name"]).size()
```

```python
hours = df["start_date_time"].unique()
stations = df["start_station_name"].unique()
data = {i:[] for i in hours}
for i in hours:
    for j in stations:
        if (i, j) not in sr.index:
            data[i].append(0)
        else:
            data[i].append(sr[i][j])
data
```

```python
df = pd.DataFrame(data).T
df.columns = stations
```

```python
df = df.sort_index()
```

```python
df.to_csv("..\dataset\data_cleaned_vis3.csv")
```

```python
station = df.rename(columns = {'long': 'x', 'lat': 'y'})[['id', 'name', 'x', 'y']]
```

```python
mapper = {}
for index, row in station.iterrows():
    mapper[row['name']] = [row['x'], row['y']]
```

```python
trip = pd.read_csv("../dataset/trip.csv")
```

```python
trip.head()
```

```python
out = []
for index, row in trip.iterrows():
    try: 
        source = {'name': row['start_station_name'], 'x': mapper[row['start_station_name']][0], 'y': mapper[row['start_station_name']][1]}
        target = {'name': row['end_station_name'], 'x': mapper[row['end_station_name']][0], 'y': mapper[row['end_station_name']][1] }
        inner = {'source': source, 'target': target}
        out.append(inner)
    except:
        continue
```

```python
import json
```

```python
with open('trip', 'w') as outfile: 
    json.dump(out, outfile)
```
